```python
import pandas as pd
from pandas_gbq import read_gbq
```

```python
query = """
SELECT *
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
LIMIT 1000
"""

df = read_gbq(query, project_id="hospital-analytics-489410")
df.head()
```

```python
query = """
SELECT *
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
"""

df = read_gbq(query, project_id="hospital-analytics-489410")
```

```python
query = """
SELECT
AVG(time_in_hospital) AS avg_length_of_stay
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
"""

los = read_gbq(query, project_id="hospital-analytics-489410")
los
```

```python
query = """
SELECT
age,
AVG(time_in_hospital) AS avg_los,
COUNT(*) AS encounters
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
GROUP BY age
ORDER BY age
"""

los_age = read_gbq(query, project_id="hospital-analytics-489410")
los_age
```

```python
query = """
SELECT
readmitted,
COUNT(*) AS encounters
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
GROUP BY readmitted
"""

readmission = read_gbq(query, project_id="hospital-analytics-489410")
readmission
```

```python
query = """
SELECT
CASE
WHEN number_inpatient > 1 OR number_emergency > 1
THEN 'High Utilizer'
ELSE 'Standard'
END AS utilization_segment,
COUNT(*) AS patients
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
GROUP BY utilization_segment
"""

utilization = read_gbq(query, project_id="hospital-analytics-489410")
utilization
```

```python
query = """
SELECT
num_medications,
AVG(time_in_hospital) AS avg_los
FROM `hospital-analytics-489410.hospital_analytics.diabetes_cleaned`
GROUP BY num_medications
ORDER BY num_medications
"""

medication = read_gbq(query, project_id="hospital-analytics-489410")
medication
```

```python
los_age.to_csv("los_by_age.csv", index=False)
readmission.to_csv("readmission_stats.csv", index=False)
utilization.to_csv("utilization_segments.csv", index=False)
medication.to_csv("medication_analysis.csv", index=False)
```
