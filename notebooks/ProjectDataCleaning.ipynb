```python
import pandas as pd

df = pd.read_csv("diabetic_data.csv")
df.head()
```

```python
df = df.replace("?", None)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=[
    "weight",
    "max_glu_serum",
    "A1Cresult"
])
```

```python
df["race"] = df["race"].fillna("Unknown")
df["payer_code"] = df["payer_code"].fillna("Unknown")
df["medical_specialty"] = df["medical_specialty"].fillna("Unknown")
```

```python
df["diag_1"] = df["diag_1"].fillna("Unknown")
df["diag_2"] = df["diag_2"].fillna("Unknown")
df["diag_3"] = df["diag_3"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df["prolonged_stay"] = (df["time_in_hospital"] >= 8).astype(int)
```

```python
df["high_utilizer"] = (
    (df["number_emergency"] > 1) |
    (df["number_inpatient"] > 1)
).astype(int)
```

```python
df["patient_complexity"] = "Low"

df.loc[df["num_medications"] > 10, "patient_complexity"] = "Moderate"
df.loc[df["num_medications"] > 20, "patient_complexity"] = "High"
```

```python
df.tail()
```

```python
df.to_csv("diabetes_cleaned.csv", index=False)
```
